```
import random
import time
import matplotlib.pyplot as plt

def linear_search(arr, key):
    for i in range(len(arr)):
        if arr[i] == key:
            return i
    return -1

# Generate random list
n = 10000
arr = random.sample(range(1, 10 * n), n)

# Define keys
key_best = arr[0]          # First element
key_average = arr[n // 2]  # Middle element
key_worst = -1             # Not in the list

# Measure Best Case Time
start = time.time()
linear_search(arr, key_best)
time_best = (time.time() - start) * 1000

# Measure Average Case Time
start = time.time()
linear_search(arr, key_average)
time_average = (time.time() - start) * 1000

# Measure Worst Case Time
start = time.time()
linear_search(arr, key_worst)
time_worst = (time.time() - start) * 1000

# Output execution times
print("Execution Times (in milliseconds):")
print(f"Best Case: {time_best:.4f} ms")
print(f"Average Case: {time_average:.4f} ms")
print(f"Worst Case: {time_worst:.4f} ms")

# Plotting
cases = ['Best Case', 'Average Case', 'Worst Case']
times = [time_best, time_average, time_worst]

plt.figure(figsize=(8, 5))
plt.bar(cases, times, color=['green', 'orange', 'red'])
plt.title('Linear Search: Case-wise Time Comparison')
plt.xlabel('Case Type')
plt.ylabel('Time (ms)')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```
