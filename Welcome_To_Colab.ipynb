```
import random
import matplotlib.pyplot as plt

# Linear search with comparison count
def linear_search_with_comparisons(arr, key):
    comparisons = 0
    for i in range(len(arr)):
        comparisons += 1
        if arr[i] == key:
            return comparisons
    return comparisons

# Input sizes to test
sizes = [100, 500, 1000, 5000, 10000]
best_comparisons = []
average_comparisons = []
worst_comparisons = []

for n in sizes:
    arr = random.sample(range(1, 10 * n), n)

    key_best = arr[0]            # Best case
    key_average = arr[n // 2]    # Average case
    key_worst = -1               # Worst case (not in array)

    best = linear_search_with_comparisons(arr, key_best)
    average = linear_search_with_comparisons(arr, key_average)
    worst = linear_search_with_comparisons(arr, key_worst)

    best_comparisons.append(best)
    average_comparisons.append(average)
    worst_comparisons.append(worst)

# Display results as table
print(f"{'Input Size':<12}{'Best Case':<12}{'Average Case':<15}{'Worst Case':<12}")
for i in range(len(sizes)):
    print(f"{sizes[i]:<12}{best_comparisons[i]:<12}{average_comparisons[i]:<15}{worst_comparisons[i]:<12}")

# Plotting
plt.figure(figsize=(10, 6))
plt.plot(sizes, best_comparisons, 'go-', label='Best Case')
plt.plot(sizes, average_comparisons, 'yo-', label='Average Case')
plt.plot(sizes, worst_comparisons, 'ro-', label='Worst Case')
plt.title('Number of Comparisons vs Input Size in Linear Search')
plt.xlabel('Input Size (n)')
plt.ylabel('Number of Comparisons')
plt.legend()
plt.grid(True, linestyle='--', alpha=0.5)
plt.show()
```
